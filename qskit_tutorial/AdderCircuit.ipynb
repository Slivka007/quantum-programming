```python
import qiskit as qk
from qiskit.circuit.library import QFT
import numpy as np
import matplotlib as mpl
from qiskit.visualization import plot_histogram
```

Circuit for addition 1+1

```python
qc_11 = qk.QuantumCircuit(4,2)

qc_11.x(0) 
qc_11.x(1) 
qc_11.barrier()

qc_11.cx(0,2)
qc_11.cx(1,2)

qc_11.ccx(0,1,3)
qc_11.barrier()

qc_11.measure(2,0) 
qc_11.measure(3,1) 

qc_11.draw('mpl')
```
